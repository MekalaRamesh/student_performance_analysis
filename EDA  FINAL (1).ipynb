```python

import numpy as np

# Assuming that your data is stored in a Pandas DataFrame called df
age = student['sciences_grade']



# Calculate the interquartile range (IQR)
q1 = np.percentile(age, 25)
q3 = np.percentile(age, 75)
iqr = q3 -q1
print("IQR",iqr)
# Calculate the values of the lower and upper whiskers
lower_whisker = q1 - 1.5 * iqr
upper_whisker = q3 + 1.5 * iqr

print('lower Whisker', lower_whisker)
print("upper Whisker",upper_whisker)
```

```python
df_filtered = student[(student['sciences_grade'] >= lower_whisker) & (student['sciences_grade'] <= upper_whisker)]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


numerical_features = [ 'age', 'english_grade', 'math_grade', 'sciences_grade', 'language_grade','portfolio_rating', 'coverletter_rating', 'refletter_rating']


for feature in numerical_features:
    plt.figure(figsize=(2, 1))
    sns.histplot(df_filtered[feature], kde=True)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# List of numerical features
numerical_features =[ 'age', 'english_grade', 'math_grade', 'sciences_grade', 'language_grade','portfolio_rating', 'coverletter_rating', 'refletter_rating']


# Determine the type of distribution
def get_distribution(data):
    skewness = stats.skew(data)
    if skewness > 0:
        return 'Right Skewed Distribution'
    elif skewness < 0:
        return 'Left Skewed Distribution'
    else:
        return 'Normal Distribution'

# Plot distribution of numerical features
fig, axs = plt.subplots(len(numerical_features)//2, 2, figsize=(15, 15))
axs = axs.ravel()
for i, feature in enumerate(numerical_features):
    sns.histplot(df_filtered[feature], kde=True, ax=axs[i])
    axs[i].set_title(f'{feature} - {get_distribution(df_filtered[feature])}')

plt.tight_layout()
plt.show()
```

## HYPOTHESIS TESTING 

```python
from scipy.stats import ttest_ind

# Separate data for male and female students
male_grades = df_filtered[df_filtered["gender"] == "M"]["math_grade"]
female_grades = df_filtered[df_filtered["gender"] == "F"]["math_grade"]

# Perform t-test
t_statistic, p_value = ttest_ind(male_grades, female_grades)
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in grades.")
else:
    print("Fail to reject the null hypothesis: No significant difference in grades.")
```
